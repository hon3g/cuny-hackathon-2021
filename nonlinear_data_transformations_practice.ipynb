```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('clean_data.csv')

df.hist(grid=True, figsize=(12,8), bins=50)
plt.show()
```

```python
sns.displot(x='projects', data=df, kde=True, bins=30)
```

```python
sns.histplot(x=df['projects'].rank(), kde=True, bins=30)
```

```python
sns.displot(x=df['projects'].rank(), kind='kde')
```

```python
sns.histplot(x=np.sqrt(df['projects']), kde=True, bins=30)
```

```python
sns.displot(x=np.log(df['projects'] + 1), kind='kde')
```

```python
sns.displot(x=np.sqrt(df['projects']), kind='kde')
```

```python
sns.displot(x=df['num_skills'], kind='kde')
```

```python
sns.displot(x=df['num_skills'])
```

```python
# log normal
sns.displot(x=np.random.lognormal(sigma=0.6, size=478), kde=True, bins=30)
```

```python
# chi square
sns.displot(x=np.random.chisquare(df=1, size=478), kde=True, bins=30)
```

```python
# num_skills
sns.displot(x=df['num_skills'], kde=True, bins=30)
```

```python
# log data
sns.displot(x=np.log(df['num_skills'] + 1), kde=True, bins=30)
```

```python
# sqrt data
sns.displot(x=np.sqrt(df['num_skills'] + 1), kde=True, bins=30, color='g')
```

```python
from scipy.stats import zscore

data = df['num_skills']
logz = zscore(np.log(df['num_skills'] + 1))
sqrtz = zscore(np.sqrt(df['num_skills'] + 1))
```

```python
plt.plot(data, 'h', markersize=3, color='y')
plt.title('data')
plt.show()
plt.plot(logz, 'h', markersize=3)
plt.title('log data')
plt.show()
plt.plot(sqrtz, 'gh', markersize=3)
plt.title('sqrt data')
plt.show()
```

```python
sns.swarmplot(y=data, size=3, color='y')
plt.show()
sns.swarmplot(y=logz, size=3)
plt.show()
sns.swarmplot(y=sqrtz, color='g', size=3)
plt.show()
```

```python
sns.boxplot(data=data, color='y')
plt.show()
sns.boxplot(data=logz)
plt.show()
sns.boxplot(data=sqrtz, color='g')
plt.show()
```

```python
plt.plot(np.log(range(1, 100)), label='log')
plt.plot(np.sqrt(range(1, 100)), label='sqrt', color='g')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
